```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
import pandas as pd
```

```python
labels_ot = ['Vehicle', 'App', 'Solar Panels']
```

```python
labels_rec = ['Fuel', 'Veh. Main', 'Driver', 'Cameras', 'App Main.', 'SP Main.']
```

```python
ot_data = [146640, 152250, 20000]
```

```python
mult_ot_data = [3*x for x in ot_data]
print(mult_ot_data)
```

```python
rec_data = [3600, 3510, 105000, 3000, 30450, 1200]
```

```python
mult_rec_data = [3*x for x in rec_data]
print(mult_rec_data)
```

```python
#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

sns.set(rc = {'figure.figsize':(15,8)})
sns.set(font_scale = 1.5)

#create pie chart
plt.pie(mult_ot_data, colors = colors)
plt.title('New York, Seattle, Austin')
plt.suptitle('Initial Costs')
plt.legend(labels_rec,bbox_to_anchor=(1.25,1), loc = 'upper right')
plt.show()
```

```python
sns.set(font_scale = 1.5)
plt.pie(mult_rec_data, colors = colors)
plt.title('New York, Seattle, Austin')
plt.suptitle('Recurring Costs (Yearly)')
plt.legend(labels_rec,bbox_to_anchor=(1.25,1), loc = 'upper right')
plt.show()
```

```python
v_type = ['Ford E-Transit','Traditional Van','Ford E-Transit','Traditional Van']
measure = ['Gas','Gas','Miles','Miles']
amount = [8.4, 76,120, 20]

sns.set(font_scale = 4)

d = {'vehicle': v_type, 'Consumption': measure, 'Amount':amount}
df = pd.DataFrame(data=d)

sns.set(rc = {'figure.figsize':(50,50)})

# Draw a nested barplot by species and sex
sns.set(font_scale = 1.5)
g = sns.catplot(
    data=df, kind="bar",
    x="Consumption", y="Amount", hue="vehicle",
    ci="sd", palette="pastel",height = 8
)
g.despine(left=True)
plt.title('Comparison of Ford E-Transit and Traditional Van Fuel and Costs', fontsize=20)
g.set(ylabel=None)
g.set(yticks=[])
g.set_axis_labels("", "")
g.legend.set_title("")
```
